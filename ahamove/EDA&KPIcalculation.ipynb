```python
import pandas as pd

# Define file paths
file_paths = {
    "Operations": "operations_full_kpi_data.csv",
    "Customer Support": "customer_support_kpi_data.csv",
    "Sales": "sales_kpi_data.csv",
    "Tech": "tech_kpi_data.csv",
    "Product": "product_kpi_data.csv",
    "Marketing": "marketing_kpi_data.csv",
    "Finance": "finance_kpi_data.csv"
}

# Iterate through each file
for name, path in file_paths.items():
    print(f" {name} Department")
    try:
        df = pd.read_csv(path)
        print(f" Rows: {df.shape[0]}, Columns: {df.shape[1]}")
        missing = df.isnull().sum()
        if missing.sum() == 0:
            print(" No missing values.\n")
        else:
            print(" Missing values found:")
            print(missing[missing > 0], "\n")
    except Exception as e:
        print(f" Error reading {path}: {e}\n")
```

```python
import pandas as pd

# Load CSVs
ops = pd.read_csv("operations_full_kpi_data.csv")
cs = pd.read_csv("customer_support_kpi_data.csv")
sales = pd.read_csv("sales_kpi_data.csv")
tech = pd.read_csv("tech_kpi_data.csv")
product = pd.read_csv("product_kpi_data.csv")
marketing = pd.read_csv("marketing_kpi_data.csv")
finance = pd.read_csv("finance_kpi_data.csv")

# ========== OPERATIONS ==========
ops_kpis = {
    "Risk-Cost Ratio (%)": (ops["unexpected_risk_cost"].sum() / (len(ops) * 20000)) * 100,

    "Commission Rate (%)": ops["commission_revenue"].sum() / ops["total_order_value"].sum() * 100,

    "Delivery Success Rate (%)": (ops["delivery_status"] == "Success").mean() * 100,

    "On-Time Completion Rate (%)": (ops["sla_met"] == "Yes").mean() * 100,

    "Driver Quality Rate (%)": (
        ops[ops["driver_rating"] >= 4.9]["driver_id"].nunique() /
        ops["driver_id"].nunique()
    ) * 100,

    "Driver Retention Rate (%)": (
        ops[(ops["active_last_month"] == 1) & (ops["active_current_month"] == 1)]["driver_id"].nunique() /
        ops[ops["active_last_month"] == 1]["driver_id"].nunique()
    ) * 100
}


# ========== CUSTOMER SUPPORT ==========
cs_kpis = {
    "Cost per Ticket (VND)": cs["support_cost"].mean(),  
    "First-Contact Resolution Rate (%)": (cs["first_contact_resolution"] == "Yes").mean() * 100,
    "CSAT Score (avg)": cs["csat_score"].mean(),  
    "SLA Fulfillment Rate (%)": (cs["sla_met"] == "Yes").mean() * 100
}

# ========== SALES ==========
sales_kpis = {
    "GSV Growth Approx. (%)": sales.groupby("date")["gsv"].sum().pct_change().mean() * 100,
    "Revenue Margin (%)": (sales["aha_revenue"] - sales["order_cost"]).sum() / sales["aha_revenue"].sum() * 100,
    "LTV/CAC Ratio": (sales["ltv"] / sales["cac"]).mean(),
    "Repeat-Customer Rate (%)": (sales["is_repeat_customer"] == 1).mean() * 100,
    "Diamond Churn Rate (%)": sales.query("customer_segment == 'Diamond'")["churned"].mean() * 100,
    "Platinum Churn Rate (%)": sales.query("customer_segment == 'Platinum'")["churned"].mean() * 100
}

# ========== TECH ==========
tech_kpis = {
    "Dev Cost vs Budget (%)": (tech["dev_cost"].sum() / tech["budget"].sum()) * 100,
    "Cloud Spend per Order (VND)": (tech["cloud_cost"] * tech["orders_processed"]).sum() / tech["orders_processed"].sum(),
    "IT Expense Ratio (%)": (tech["it_expense"].sum() / tech["total_revenue"].sum()) * 100,
    "System Uptime (%)": (tech["uptime_minutes"].sum() / (tech["scheduled_minutes"].sum())) * 100,
    "MTTR (min)": (tech["downtime_minutes"].sum() / tech["incident_count"].sum()),
    "Internal NPS (avg)": tech["internal_nps"].mean()
}

# ========== PRODUCT ==========
product_kpis = {
    "ROI per Release (%)": (product["revenue_from_release"] / product["dev_cost"]).mean() * 100,
    "Cost per Feature (VND)": product["cost_per_feature"].mean(),
    "Time to Market (days)": product["time_to_market_days"].mean(),
    "Feature Adoption Rate (%)": product["feature_adoption_rate"].mean(),
    "User Retention Rate (%)": product["user_retention_rate"].mean(),
    "In-App NPS (avg)": product["inapp_nps"].mean()
}

# ========== MARKETING ==========
marketing_kpis = {
    "CAC (VND)": (marketing["marketing_spend"] / marketing["new_customers"]).mean(),
    "Marketing Spend Effectiveness": (marketing["revenue_attributed"] / marketing["marketing_spend"]).mean(),
    "Budget Variance (%)": ((marketing["actual_spend"] / marketing["budgeted_spend"]).mean() - 1) * 100,
    "Brand Awareness Index (avg)": marketing["brand_awareness_index"].mean(),
    "Lead Quality Score (avg)": marketing["lead_quality_score"].mean(),
    "Conversion Rate (%)": marketing["conversion_rate"].mean()
}

# ========== FINANCE ==========
finance_kpis = {
    "Net Profit Margin (%)": (finance["net_profit"].sum() / finance["revenue"].sum()) * 100,
    "Operating Cash Flow (avg)": finance["operating_cash_flow"].mean(),
    "Days Sales Outstanding (avg)": finance["dso"].mean(),
    "Budget Accuracy Score (%)": finance["budget_accuracy_score"].mean(),
    "Internal CSAT (avg)": finance["internal_csat_score"].mean()
}

# Display
print(" OPERATIONS KPIs:\n", ops_kpis)
print("\n CUSTOMER SUPPORT KPIs:\n", cs_kpis)
print("\n SALES KPIs:\n", sales_kpis)
print("\n TECH KPIs:\n", tech_kpis)
print("\n PRODUCT KPIs:\n", product_kpis)
print("\n MARKETING KPIs:\n", marketing_kpis)
print("\n FINANCE KPIs:\n", finance_kpis)
```
